```python
import pandas as pd
```

```python
# Import the data as a Pandas DataFrame. Name the DataFrame bike_df. 
# Then, complete the following:

# Delete the unnecessary columns:
# Unnamed: 7
# Unnamed: 8
# Unnamed: 9

bike_df = pd.read_excel('/content/RailsToTrails_National Count Data_week.xlsx')
print(bike_df.keys()) # This line added in order to copy unwanted column names

bike_df = bike_df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])
bike_df.head()

```

```python
# 2. Rename the Unnamed: 6 column to percent_change
bike_df = bike_df.rename(columns={'Unnamed: 6': 'percent_change'})
bike_df.head()
```

```python
# 3. How many missing values are in the dataset? Which columns are they in?
bike_df.isna().sum()
# Answer: There are 52 missing values in the 2021 counts column
```

```python
# Fill all missing values in the 2021 Counts column with 0.
bike_df['2021 Counts'].fillna(0, inplace=True)
bike_df.isna().sum() # to confirm that data set has 0 missing values
```

```python
# Optional/Bonus: There is a challenging but common issue in the name for 
# the 2019 counts column. Can you find it?
# If so, Rename the 2019 counts 
# (31 counters) column to "counts_2019" (a more Pythonic column name)

bike_df = bike_df.rename(columns={' 2019 counts (31 counters)': 'counts_2019',
                        ' 2020 counts (31 counters)': 'counts_2020',
                        '2021 Counts': 'counts_2021'
                        })                       

bike_df.head()

```

Answer: It is all of the spaces in the column name, it makes it harder to use certian methods/code. For example you can't look at the series by using the dot method ie: df.column name <- will populate error because of the space.
